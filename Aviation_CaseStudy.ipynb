# **Step-4- Data Cleaning**



```python
aviation_accidents_df=aviation_accidents.copy() #Before cleaning the data creating a new data frame is always a good practice.
```

```python
aviation_accidents_df = aviation_accidents_df.apply(lambda x: x.str.strip() if x.dtype == 'object' else x) #stripoff all the empty white spaces before starting anything
```

```python
aviation_accidents_df['date'] = pd.to_datetime(aviation_accidents_df['date'], errors='coerce') # to change the datatype of the date
```

Observed few inconsistencies in fatilites like 8+1, first add up all the data and then convert the datatype

```python
#To add up the fatilites like 8+1 and manage the inconsistency
aviation_accidents_df['fatilites'] = aviation_accidents_df['fatilites'].apply(lambda x: eval(x) if isinstance(x, str) and '+' in x and x.replace('+','').replace(' ','').isdigit() else x)
#To change the data type from object to Int
aviation_accidents_df['fatilites'] = pd.to_numeric(aviation_accidents_df['fatilites'], errors='coerce').astype('Int64') #Int64 is a nullable integer dtype
```

One other thing that can be done is handling the fatilites aas it has the highest percentage of missing values i.e - 14.10

To handle that I think the category of hull-loss and repairable damage can be used. What can be done is if the value of fatilites is NaN then we can parse the effect column if the effect is "1" then we can assume there might me fatilites if the effect is 0 we can assume as 0, its just an assumption based on my understanding


```python
hull_loss_median = aviation_accidents_df[
    (aviation_accidents_df['effect'] == '1') &
    (aviation_accidents_df['fatilites'].notna())
]['fatilites'].median()

aviation_accidents_df['fatilites'] = aviation_accidents_df.apply(
    lambda row: row['fatilites'] if pd.notna(row['fatilites'])
    else (hull_loss_median if row['effect'] == '1' else 0),
    axis=1
)
```

Before printing the visualizations, I want to create a callable function for the colour palette I like to use, the colour blindness free palette, just for everyone's convenience and aesthetics
"https://immunobiology.duke.edu/sites/default/files/2023-04/Colorblind-Palette.pdf"

```python
def colour_palette(n_colors=15):
    colors = [
        # Bright
        '#4477AA', '#66CCEE', '#228833', '#CCBB44', '#EE6677', '#AA3377', '#BBBBBB',
        # Vibrant
        '#0077BB', '#33BBEE', '#009988', '#EE7733', '#CC3311', '#EE3377',
        # Muted
        '#332288', '#88CCEE', '#44AA99', '#117733', '#999933', '#DDCC77',
        '#CC6677', '#882255', '#AA4499', '#DDDDDD'
    ]

    return colors[:n_colors]
```

#**Chart-1**

Tried to plot the pie chart using matplotlib but it turned out to be messy so changed to plotly by following - "https://plotly.com/python/pie-charts/"

```python
import plotly.express as px

# Calculate total for the other countries
total_accidents_all = accidents_by_country['Total_Accidents'].sum()
top15_total = top15_countries['Total_Accidents'].sum()
other_accidents = total_accidents_all - top15_total

# Create data for pie chart
pie_data = top15_countries[['Country', 'Total_Accidents']].copy()
other_row = pd.DataFrame({'Country': ['Other'], 'Total_Accidents': [other_accidents]})
pie_data = pd.concat([pie_data, other_row], ignore_index=True)

# Calculate percentages
pie_data['Percentage'] = (pie_data['Total_Accidents'] / total_accidents_all) * 100

# Get colors
colors = colour_palette(16)

# Create Plotly pie chart
fig = px.pie(pie_data,
             values='Total_Accidents',
             names='Country',
             color_discrete_sequence=colors,
             hole=0.3)  # Makes it a donut chart - cleaner look

fig.update_traces(
    textposition='outside',
    textinfo='label+percent',
    pull=[0.1 if i == 0 else 0 for i in range(len(pie_data))]  # Pull out USA slightly
)

fig.update_layout(
    title={
        'text': 'Percentage Distribution of Aviation Accidents (Top 15 + Other)',
        'font': {'size': 16, 'family': 'Arial', 'color': 'black'},
        'x': 0.5,
        'xanchor': 'center'
    },
    showlegend=True,
    height=700,
    width=1000
)

fig.show()

# Summary
print(f"Top 15 countries represent: {(top15_total/total_accidents_all)*100:.2f}% of total accidents")
print(f"Other countries represent: {(other_accidents/total_accidents_all)*100:.2f}% of total accidents")
```

```python
import plotly.graph_objects as go #https://plotly.com/python/graph-objects/ followed the documentation

#get all the countries
top15 = top15_countries.copy()
others = accidents_by_country[~accidents_by_country['Country'].isin(top15['Country'])].sort_values('Total_Accidents', ascending=False)

# Get colors
colors = colour_palette(16)

fig = go.Figure()

# individual bars for top 15 countries
for i in range(len(top15)):
    fig.add_trace(go.Bar(
        x=[top15.iloc[i]['Country']],
        y=[top15.iloc[i]['Total_Accidents']],
        name=top15.iloc[i]['Country'],
        marker_color=colors[i],
        hovertemplate=f"<b>{top15.iloc[i]['Country']}</b><br>Accidents: %{{y}}<extra></extra>"
    ))

# stacked into one bar for others
for idx, row in others.iterrows():
    fig.add_trace(go.Bar(
        x=['Other Countries'],
        y=[row['Total_Accidents']],
        name=row['Country'],
        marker_color=colors[15],
        hovertemplate=f"<b>{row['Country']}</b><br>Accidents: {row['Total_Accidents']}<extra></extra>",
        showlegend=False
    ))

fig.update_layout(
    barmode='stack',
    title={
        'text': 'Chart 1.0: Total Number of Accidents by Country (Top 15 + Other)',
        'font': {'size': 16, 'family': 'Arial, sans-serif', 'color': 'black'},
        'x': 0.5,
        'xanchor': 'center',
        'y': 0.98,
        'yanchor': 'top'
    },
    xaxis_title='Country',
    yaxis_title='Total Accidents',
    height=700,
    width=1400,
    font=dict(weight='bold')
)

fig.show()
```

# **Chart 2- Global Map**

To get a basic understanding of the globe plotting saw a Youtube video - "https://youtu.be/EvRpmoMTXlk?si=VHv99HJD2U2e7UOB" and then read the documentation - "https://plotly.com/python/bubble-maps/"

```python
# Create country map from existing dataframe created before
country_mapping = pd.DataFrame({
    'Country': country_names,
    'Country_Code': country_codes
})

# Merge with accidents_by_country
accidents_map = accidents_by_country.merge(country_mapping, on='Country', how='left')

# Remove any invalid country codes
accidents_map = accidents_map[accidents_map['Country_Code'] != 'not found']

# Get colors from the precreated function
colors = colour_palette(len(accidents_map))

# Create the map
fig = px.scatter_geo(accidents_map,
                     locations='Country_Code',
                     locationmode='ISO-3',
                     size='Total_Accidents',
                     color='Country',
                     hover_name='Country',
                     hover_data={
                         'Country_Code': True,
                         'Total_Accidents': True,
                         'Country': False
                     },
                     color_discrete_sequence=colors,
                     projection='natural earth')

# layout
fig.update_layout(
    title={
        'text': 'Chart 2.0: Global Distribution of Aviation Accidents by Country',
        'font': {
            'size': 16,
            'color': 'black',
            'family': 'Arial, sans-serif',
            'weight': 'bold'
        },
        'x': 0.5,
        'xanchor': 'center',
    },
    geo=dict(
        showframe=True,
        showcoastlines=True,
        projection_type='natural earth'
    ),
    showlegend=True,
    legend=dict(
        orientation="v",
        yanchor="middle",
        y=0.5,
        xanchor="left",
        x=1.02
    ),
    height=700,
    width=1400
)

fig.show()
```

# **Chart 4 - Heat Map**

```python
# Create crosstab of cause and effect
heatmap_data = pd.crosstab(
    aviation_accidents_df['cause'],
    aviation_accidents_df['effect'],
    margins=False
)

# Map the codes to readable labels
cause_labels = {
    'A': 'Accident',
    'H': 'Hijacking',
    'O': 'Other Occurrence',
    'C': 'Criminal',
    'U': 'Unknown',
    'I': 'Incident'
}

effect_labels = {
    '1': 'Hull-loss',
    '2': 'Repairable Damage'
}

# Rename rows and columns
heatmap_data.index = heatmap_data.index.map(lambda x: cause_labels.get(x, x))
heatmap_data.columns = heatmap_data.columns.map(lambda x: effect_labels.get(x, x))

fig = px.imshow(
    heatmap_data,
    labels=dict(x="Damage Type", y="Cause", color="Accident Count"),
    x=heatmap_data.columns,
    y=heatmap_data.index,
    color_continuous_scale='YlOrRd',  #
    text_auto=True,
    aspect='auto'
)

fig.update_layout(
    title={
        'text': '<b>Chart 4.0: Heatmap of Accident Cause vs Damage Severity</b>',
        'font': {'size': 18, 'family': 'Arial', 'color': 'black'},
        'x': 0.5,
        'xanchor': 'center',
        'y': 0.95,
        'yanchor': 'top'
    },
    xaxis_title="Damage Type (Effect)",
    yaxis_title="Accident Cause",
    height=600,
    width=900,
    font=dict(size=12)
)

fig.update_traces(
    texttemplate='<b>%{z}</b>',
    textfont={"size": 14}
)

fig.show()
```

```python
# Quick percentage breakdown
crosstab_pct = pd.crosstab(
    aviation_accidents_df['cause'],
    aviation_accidents_df['effect'],
    normalize='index'
) * 100

print("Percentage of each cause by damage type:")
print(crosstab_pct.round(1))
```
